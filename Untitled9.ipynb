```python
from ultralytics import YOLO
```

```python
yaml_content = """path: /content/pole2.0
train: images/train
val: images/val

nc: 9
names:
  0: Utility pole
  1: Telecom pole
  2: Roof top tower
  3: Crossarms
  4: Insulators
  5: Light
  6: Transformers
  7: Wire
  8: Telecom equipment
"""

with open("/content/pole2.0/data.yaml", "w") as f:
    f.write(yaml_content)

print("✅ YAML saved to /content/pole2.0/data.yaml")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")

results = model.train(
    data="/content/pole2.0/data.yaml",
    epochs=150,
    patience=50,
    batch=16,
    imgsz=640,
    optimizer="AdamW",
    lr0=0.01,
    lrf=0.01,
    warmup_epochs=5,
    cos_lr=True,
    weight_decay=0.0005,
    mosaic=1.0,
    mixup=0.1,
    degrees=10.0,
    scale=0.5,
    translate=0.1,
    fliplr=0.5,
    flipud=0.0,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    close_mosaic=20,
    project="pole_detection",
    name="yolov8m_poles_v1",
    save=True,
    save_period=25,
    plots=True,
    verbose=True,
)

print("\n✅ TRAINING COMPLETE!")
print("Best weights: pole_detection/yolov8m_poles_v1/weights/best.pt")
```

```python
import gradio as gr
from ultralytics import YOLO
from PIL import Image

model = YOLO("/content/runs/detect/pole_detection/yolov8m_poles_v1/weights/best.pt")

def detect(image, confidence):
    results = model.predict(source=image, conf=confidence, iou=0.45)
    plotted = results[0].plot()
    return Image.fromarray(plotted[..., ::-1])

demo = gr.Interface(
    fn=detect,
    inputs=[
        gr.Image(type="pil", label="Upload Image"),
        gr.Slider(0.1, 0.9, value=0.25, step=0.05, label="Confidence Threshold"),
    ],
    outputs=gr.Image(type="pil", label="Detection Result"),
    title="Pole Detection - YOLOv8m",
    description="Upload an image to detect Utility poles, Telecom poles, Roof top towers and other components.",
)

demo.launch(share=True)
```
